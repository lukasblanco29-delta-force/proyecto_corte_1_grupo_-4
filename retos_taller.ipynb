```python
# reto3_dolar_track.py

trm_semana = []

# Menu interactivo
while True:
    print("\n=== Dolar-Track ===")
    print("1. Ingresar TRM de 7 dias")
    print("2. Analizar semana")
    print("3. Ver TRM ordenada")
    print("4. Salir")
    opcion = input("Opcion: ")

    if opcion == "1":
        trm_semana = []
        for i in range(7):
            while True:
                try:
                    valor = float(input(f"TRM dia {i+1}: "))
                    if valor <= 0:
                        print("Error: la TRM debe ser mayor a 0.")
                    else:
                        trm_semana.append(valor)
                        break
                except ValueError:
                    print("Ingresa un numero valido.")
        print("TRM registrada.")

    elif opcion == "2":
        if len(trm_semana) < 7:
            print("Primero ingresa los 7 valores de TRM.")
            continue

        promedio = sum(trm_semana) / len(trm_semana)

        arriba = 0
        abajo = 0
        for valor in trm_semana:
            if valor > promedio:
                arriba += 1
            elif valor < promedio:
                abajo += 1

        volatilidad = max(trm_semana) - min(trm_semana)

        print(f"\nPromedio semanal: {promedio:.2f}")
        print(f"Dias por encima del promedio (Alerta Venta): {arriba}")
        print(f"Dias por debajo del promedio (Alerta Compra): {abajo}")
        print(f"Volatilidad semanal (max-min): {volatilidad:.2f}")

    elif opcion == "3":
        if len(trm_semana) < 7:
            print("Primero ingresa los 7 valores de TRM.")
            continue
        ordenada = trm_semana.copy()
        ordenada.sort()
        print("TRM ordenada (menor a mayor):", ordenada)

    elif opcion == "4":
        print("Saliendo de Dolar-Track.")
        break

    else:
        print("Opcion invalida.")
```

# A continuacion reto Administracion

```python
# reto5_nomina_express.py

VALOR_HORA = 10000.0
AUX_TRANSPORTE = 162000.0
SMMLV = 1300000.0  # simulado
TOPE_AUX = 2 * SMMLV

nombres = []
horas = []

while True:
    print("\n=== Nomina-Express ===")
    print("1. Ingresar 5 empleados")
    print("2. Calcular nomina")
    print("3. Ver empleados ordenados por nombre")
    print("4. Salir")
    opcion = input("Opcion: ")

    if opcion == "1":
        nombres = []
        horas = []
        for i in range(5):
            nombre = input(f"Nombre empleado #{i+1}: ").strip()
            while True:
                try:
                    h = float(input("Horas trabajadas: "))
                    if h < 0:
                        print("Error: las horas no pueden ser negativas.")
                    else:
                        nombres.append(nombre)
                        horas.append(h)
                        break
                except ValueError:
                    print("Ingresa un numero valido.")
        print("Datos registrados.")

    elif opcion == "2":
        if len(nombres) < 5:
            print("Primero debes ingresar los 5 empleados.")
            continue

        total_empresa = 0.0
        print("\n--- Desprendible de pago ---")

        for i in range(5):
            h = horas[i]

            if h > 48:
                ordinarias = 48
                extras = h - 48
            else:
                ordinarias = h
                extras = 0

            pago_ordinario = ordinarias * VALOR_HORA
            pago_extra = extras * (VALOR_HORA * 1.25)
            subtotal = pago_ordinario + pago_extra

            if subtotal < TOPE_AUX:
                aux = AUX_TRANSPORTE
            else:
                aux = 0.0

            total_empleado = subtotal + aux
            total_empresa += total_empleado

            print(f"\nEmpleado: {nombres[i]}")
            print(f"  Horas: {h}")
            print(f"  Pago ordinario: ${pago_ordinario:,.0f}")
            print(f"  Pago extras: ${pago_extra:,.0f}")
            print(f"  Auxilio transporte: ${aux:,.0f}")
            print(f"  Total a pagar: ${total_empleado:,.0f}")

        print(f"\nTOTAL NOMINA EMPRESA: ${total_empresa:,.0f}")

    elif opcion == "3":
        if len(nombres) < 5:
            print("Primero debes ingresar los 5 empleados.")
            continue
        data = list(zip(nombres, horas))
        data.sort(key=lambda x: x[0].lower())
        print("\nEmpleados ordenados:")
        for n, h in data:
            print(f"{n} - {h} horas")

    elif opcion == "4":
        print("Saliendo de Nomina-Express.")
        break

    else:
        print("Opcion invalida.")
```
